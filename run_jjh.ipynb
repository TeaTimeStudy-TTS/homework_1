```python
from selenium import webdriver as wd
from selenium.webdriver.common.by import By
import pandas as pd
```

```python
# 자동 제어 브라우저 가동
driver = wd.Chrome()
```

```python
# 타겟 사이트 진입
driver.get('https://www.starbucks.co.kr/store/store_map.do')
driver.implicitly_wait(10)
```

## 수집정보 
- 서울시 스타벅스 매장정보
  - 매장명
  - 주소
  - 매장타입(general, reserve, generalDT)
  
## 저장방식
- csv

# 진입순서
1. 지역검색 클릭
2. 서울 클릭
3. 전체 클릭
(화면 깜빡임)
4. 데이터 수집 시작
5. csv 출력

```python

region_click_css = 'header.loca_search > h3 > a'
driver.find_element(By.CSS_SELECTOR, region_click_css).click()
```

```python
seoul_click_css = 'div.loca_step1_cont > ul > li:nth-child(1) > a'
driver.find_element(By.CSS_SELECTOR, seoul_click_css).click()
```

```python
all_click_css = '#mCSB_2_container > ul > li:nth-child(1) > a'
driver.find_element(By.CSS_SELECTOR, all_click_css).click()

driver.implicitly_wait(5)
```

```python
stores = driver.find_elements(By.CSS_SELECTOR, '#mCSB_3_container > ul > li')
stores_name = [ store.get_attribute('data-name') for store in stores]
```

```python
addresses = driver.find_elements(By.CSS_SELECTOR, '#mCSB_3_container > ul > li > p')
addresses_name = [ address.get_attribute("textContent").replace('1522-3232', '').strip() for address in addresses]
```

```python
types = driver.find_elements(By.CSS_SELECTOR, '#mCSB_3_container > ul > li > i') 
types_name = [ type.get_attribute('class').replace('pin_', '').strip() for type in types ]
```

```python
columns = ['매장명', '매장주소', '매장타입']
datas = {
  '매장명' : stores_name,
  '매장주소' : addresses_name,
  '매장타입' : types_name
}
df = pd.DataFrame(datas)
df.sample(10)
df.to_csv('장정현.csv', index=False, encoding='utf-8')
```

```python
driver.quit()
```
